# Train Test Split

```python
from sklearn.model_selection import train_test_split

X = dfinals.drop(['product_wg_ton'], axis = 1)
Y = dfinals['product_wg_ton']

X_train, X_test, Y_train , Y_test = train_test_split(X,Y,train_size = 0.7 , test_size = 0.3, random_state = 0)
```

```python
X_test.isnull().sum()
```

# Feature Scaling

```python
# Using Min max scaler for Better Accuracy
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()


scaler.fit(X_train)


X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_test_scaled.describe()
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr_scaled = LinearRegression()
lr.fit(X_train, Y_train)
lr_scaled.fit(X_train_scaled, Y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_absolute_percentage_error

print("R2_Score :", r2_score(Y_test, y_pred_scaled))
print("*****************************")
print("MAE :", mean_absolute_error(Y_test, y_pred_scaled))
print("*****************************")
print("MAPE :", mean_absolute_percentage_error(Y_test, y_pred_scaled))
```

```python
plt.scatter(Y_test, y_pred_scaled)
plt.show
```

```python
from sklearn.tree import DecisionTreeRegressor
Dt= DecisionTreeRegressor()
dt_scaled = DecisionTreeRegressor()
Dt.fit(X_train ,Y_train)
dt_scaled.fit(X_train_scaled, Y_train)
```

```python
y_pred = Dt.predict(X_test)
y_pred_scaled = dt_scaled.predict(X_test_scaled)
```

```python
print("R2_Score :", r2_score(Y_test, y_pred_scaled))
print("*****************************")
print("MAE :", mean_absolute_error(Y_test, y_pred_scaled))
print("*****************************")
print("MAPE :", mean_absolute_percentage_error(Y_test, y_pred_scaled))
```

```python
#Importance of feature
feat_imp = pd.Series(dt_scaled.feature_importances_, list(X_train)).sort_values(ascending=False)
fig = plt.figure(figsize=(12, 4))
feat_imp.plot(kind='bar', title='Importance of Features')
plt.ylabel('Feature Importance Score')
plt.grid()
plt.show()
```
